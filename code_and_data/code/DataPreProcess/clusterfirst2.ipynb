```python
import numpy as np
import pandas as pd
import os
```

```python
rootdir = r"../../Final_factor_287/"
savebasedir = r"../../Final_factor/"
savedir = r"../../Final_factor/all_save_data_II/"
# basedir = os.path.join(r"../../data/ExtraData","weather_by_area")
# basedir2 = os.path.join(r"../../data/ExtraData","379_area_New_cases_csv")
basedir = os.path.join(r"../../Final_factor","weather_with_cases")
age = os.path.join(rootdir,"age65.xlsx")
airpassenger = os.path.join(rootdir,"airpassenger.CSV")
doctor = os.path.join(savebasedir,"doctor_per1000.csv")
nonwhite = os.path.join(savebasedir,"non_white_292.csv")
popdens = os.path.join(savebasedir,"popdensnew.CSV")
uv = os.path.join(rootdir,"UVprocessed.csv")
```

```python
agedf = pd.read_excel(age,usecols=["LAD20CD","LAD20NM","male","female"])
airpassengerdf = pd.read_csv(airpassenger)
doctordf = pd.read_csv(doctor,usecols=["LAD20CD","LAD20NM","doctors"])
# nonwhitedf = pd.read_csv(nonwhite,usecols=["LAD20NM","nonwhite"])
popdensdf = pd.read_csv(popdens,usecols=["LAD20NM","PopulationDensity"])
uvdf = pd.read_csv(uv)
```

```python
areadf = pd.merge(agedf,popdensdf,on="LAD20NM",how="inner")
```

```python
areadf = pd.read_csv(os.path.join(rootdir,"areadf_I.csv"),index_col="LAD20NM",usecols=["LAD20NM","male","female","PopulationDensity"])
timedf = pd.read_csv(os.path.join(rootdir,"timedfnew.csv"))
```

```python
areanames = []
for base in os.listdir(basedir):
    baseroot = os.path.join(basedir,base)
    areaname = base[:-4]
    # basedf = pd.read_csv(baseroot,usecols=["Date","Temperature (°F)","Humidity (%)","Wind Speed (mph)","Pressure (inHg)"],index_col="Date")
    # try:
    #     basedf2= pd.read_csv(caseroot,usecols=["date","R0orig","R0alp","R0dlt","R0omi"])
    # except FileNotFoundError:
    #     areanames.append(areaname)
    #     continue
    basedf = pd.read_csv(baseroot,usecols=["Date","Temperature (°F)","Humidity (%)","Wind Speed (mph)","Pressure (inHg)","R0orig","R0alp","R0dlt","R0omi"],index_col="Date")
    # pcadf  = pd.read_csv(pcaroot,usecols=["date","PCAresult"])
    # pcadf.rename(columns={'date':'Date'},inplace=True)
    # df1 = basedf.merge(pcadf,how="left",on="Date")
    df2 = basedf.merge(timedf,on="Date",how="left")
    
    if areaname not in areadf.index:
        continue
    areanames.append(areaname)
    statdata = areadf.loc[areaname]
    basenp = np.ones((len(df2),3))
    for i in range(3):
        basenp[:,i]*=statdata[i]
    # extradf = pd.DataFrame(basenp,columns=["male","female","doctors","nonwhite","PopulationDensity"])
    extradf = pd.DataFrame(basenp,columns=["male","female","PopulationDensity"])
    extradf["Date"]=df2["Date"]
    df3 = pd.merge(df2,extradf,on="Date",how="left")
    df3.to_csv(os.path.join(savedir,areaname+".csv"),index=None)
    print(areaname,"Saved!")
```

```python
areadf = pd.DataFrame(areanames)
areadf.to_csv(os.path.join(savebasedir,"filenames.csv"),index=None)
```
