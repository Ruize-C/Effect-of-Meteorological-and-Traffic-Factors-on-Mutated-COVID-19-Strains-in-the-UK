```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/death.csv",usecols=["date","newDeaths28DaysByDeathDate","cumDeaths28DaysByDeathDate"],date_parser="date")
df
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
weekdf = df.resample('w',on="date").sum()
```

```python
weekdf
```

```python
for i in range(len(weekdf["cumDeaths28DaysByDeathDate"])):
    weekdf["cumDeaths28DaysByDeathDate"][i] = np.sum(weekdf["newDeaths28DaysByDeathDate"][:i+1])
```

```python
weekdf
```

```python
weekdf.to_csv("../processed/weeklydeathclustered.csv")
```
