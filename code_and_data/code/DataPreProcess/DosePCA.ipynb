```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import os
```

```python
rootdir = r"../../Final_factor_287/"
firstdir = os.path.join(rootdir,"area_First_dose")
seconddir = os.path.join(rootdir,"area_Second_dose")
thirddir = os.path.join(rootdir,"area_Third_dose")
savedir = os.path.join(rootdir,"area_PCA_dose")
```

```python
filenames = os.listdir(firstdir)
for file in filenames:
    firstdf = pd.read_csv(os.path.join(firstdir,file),usecols=["areaCode","areaName","date","cumPeopleVaccinatedFirstDoseByVaccinationDate"],index_col="date")
    seconddf = pd.read_csv(os.path.join(seconddir,file),usecols=["date","cumPeopleVaccinatedSecondDoseByVaccinationDate"],index_col="date")
    thirdddf = pd.read_csv(os.path.join(thirddir,file),usecols=["date","cumPeopleVaccinatedThirdInjectionByVaccinationDate"],index_col="date")
    alldf = firstdf.merge(seconddf.merge(thirdddf,on="date",how="left"),on="date",how="left")
    alldf.fillna(0,inplace=True)
    pca = PCA(1)
    newx = pca.fit_transform(alldf[["cumPeopleVaccinatedFirstDoseByVaccinationDate","cumPeopleVaccinatedSecondDoseByVaccinationDate","cumPeopleVaccinatedThirdInjectionByVaccinationDate"]])
    newdf = pd.DataFrame(index=firstdf.index)
    newdf["areaCode"] = firstdf["areaCode"]
    newdf["areaName"] = firstdf["areaName"]
    newdf["PCAresult"] = newx
    newdf.to_csv(os.path.join(savedir,file[:-4]+".csv"))
    print(file[:-4],"Saved!")
```
