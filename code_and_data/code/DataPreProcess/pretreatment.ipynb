### Data preprocessing code
The downloaded data on the different variants will need to be pre-merged due to the same dates and different locations, and the merged data will be the **differential** data, which will then be summed up to get the summation data.
```
Note that the data downloaded here are daily sampling data from various locations and are not representative of the whole country and need to be multiplied by a factor based on the national incidence
```

````python
import pandas as pd
import numpy as np
import time
from datetime import date
import csv
````

````python
#Here you change the name of the file to be read in, and the name of the column to be read in.
prealphadata = pd.read_csv("../data/omicron.CSV",usecols=["Date","Region","Number of Omicron cases"])
prealphadata
````

````python
alphanp = np.zeros(600,dtype=int)       #As the number of days gets larger, the 100 here may have to be correspondingly larger.
for idx, row in prealphadata.iterrows():
    d,m,y = [int(i) for i in row[0].split('/')]
    alphanp[(date(y,m,d)-date(2020,10,1)).days]+=row[2]     #Merge data with the same date
````

````python
sumalpha = np.zeros(600,dtype=int)
s = 0
for i in range(len(alphanp)):
    s+=alphanp[i]
    sumalpha[i]=s                               #totalization
sumalpha
````

````python
#Write file, file name needs to be changed
alphdf = pd.DataFrame(alphanp)
alphdf.to_csv("../processed/diff_omicron.csv",index=None)
sumdf = pd.DataFrame(sumalpha)
sumdf.to_csv("../processed/sum_omicron.csv",index=None)
````
